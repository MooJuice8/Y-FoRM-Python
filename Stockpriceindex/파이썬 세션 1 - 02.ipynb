```python
import requests

# request 내의 get 함수를 이용하여 네이버 서버를 가져오고, 이를 response라는 변수에 저장
response = requests.get("https://www.naver.com")

# .text로 response 내에 있는 소스 코드 불러옴
html = response.text 

print(html)
```

```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.naver.com")
html = response.text
soup = BeautifulSoup(html, 'lxml')
word = soup.select_one("#NM_set_home_btn")
word.text
print(word.text)
```

```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.naver.com")
html = response.text
soup = BeautifulSoup(html, 'lxml')

# id는 앞에 ‘#’을 붙여야 함, id값이 NM~인 태그 하나를 찾아냄.
word = soup.select_one("#NM_set_home_btn")

# word에서 텍스트 요소만 출력함
print(word.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://finance.naver.com/item/sise.naver?code=005930"
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'lxml')
price = soup.select_one("#_nowVal").text
price = price.replace(',', '')
print(price)
```

```python
import requests
from bs4 import BeautifulSoup
codes = ["005930", "000660", "035720"]

for code in codes:
    url = f"https://finance.naver.com/item/sise.naver?code={code}"
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'lxml')
    price = soup.select_one("#_nowVal").text
    price = price.replace(',', '')
    print(price)

```
